```python
def display_board(board):
    for row_index,row in enumerate(board):
        print("-------------")
        print("| ", end="")
        for col_index,col in enumerate(row):
            print(board[row_index][col_index] + " | ", end="")
        print("\r")
    print("-------------")
```

```python
def choose_players(players):
    players["player_1"]["name"] = input("Please enter name for player 1 : ")
    symbol = ""
    while symbol not in ["X", "O"]:
        symbol = input("{} please choose your symbol ('X' or 'O') : ".format(players["player_1"]["name"])).upper()
    players["player_1"]["symbol"] = symbol
    players["player_2"]["name"] = input("Please enter name for player 2 : ")
    if players["player_1"]["symbol"] == "X":
        players["player_2"]["symbol"] = "O"
    else:
        players["player_2"]["symbol"] = "X"
    return players
```

```python
def check_winner(board, player):
    symbol = player["symbol"]
    return ((board[0][0] == board[0][1] == board[0][2] == symbol) or
            (board[1][0] == board[1][1] == board[1][2] == symbol) or
            (board[2][0] == board[2][1] == board[2][2] == symbol) or
            (board[0][0] == board[1][0] == board[2][0] == symbol) or
            (board[0][1] == board[1][1] == board[2][1] == symbol) or
            (board[0][2] == board[1][2] == board[2][2] == symbol) or
            (board[0][0] == board[1][1] == board[2][2] == symbol) or
            (board[0][2] == board[1][1] == board[2][0] == symbol))
```

```python
def play_move(board, player):
    print("{} it's your chance.".format(player["name"]))
    symbol = player["symbol"]
    while True:
        row = -1
        col = -1
        while row not in range(1,4):
            row = input("Please enter the row (1-3) : ")
            try:
                row = int(row)
            except ValueError:
                print("Please enter a valid number for row!")
        while col not in range(1,4):
            col = input("Please enter the col (1-3) : ")
            try:
                col = int(col)
            except ValueError:
                print("Please enter a valid number for column!")
        print(board[row-1][col-1])
        if board[row-1][col-1] == " ":
            board[row-1][col-1] = symbol
            break
        else:
            print("Please choose an empty space for your move.")
            continue
    return board
```

```python
def play_game():
    while True:
        # Initialise the board
        empty_row = [' ', ' ', ' ']
        board = [ list(empty_row) for n in range(3) ]
        # Initialise the players     
        players = { "player_1" : { "name" : None, "symbol" : None }, "player_2" : { "name" : None, "symbol" : None } }
        players = choose_players(players)
        
        for chance in range(9):
            if chance % 2 == 0:
                board = play_move(board, players["player_1"])
                display_board(board)
                if check_winner(board, players["player_1"]):
                    print("{} wins!".format(players["player_1"]["name"]))
                    break
                elif chance == 8:
                    print("It's a tie!")
                    break
            elif chance % 2 == 1:
                board = play_move(board, players["player_2"])
                display_board(board)
                if check_winner(board, players["player_2"]):
                    print("{} wins!".format(players["player_2"]["name"]))
                    break
        choice = input("Do you wish to compete again? (Y/N)").upper()
        if choice == "N":
            break
```

```python
play_game()
```
